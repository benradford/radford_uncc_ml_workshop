```python
##
## In this block, we import modules (software libraries)
## and set a random seed for numpy (1234).
##
%matplotlib inline
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1234)
```

```python
##
## Define a dataset (x,y) and plot it.
## 
x = np.linspace(0,10,75)
y = np.sin(x*3) + x

f = plt.figure()
plt.plot(x, y, 'o')
```

```python
## Define a shallow, linear neural network
## called `model`.
model = Sequential([
    Dense(1, input_shape=(1,)),
    Activation('linear')
])

## Compile the model to prepare it for training.
model.compile(loss='mean_squared_error',
              optimizer="nadam",
              metrics=['accuracy'])

## Fit the model to data (x, y).
model.fit(x, y,
          epochs=20000,
          batch_size=128,
          verbose=False)

## Plot the data and the model predictions.
f = plt.figure()
plt.scatter(x,y,facecolors="none",edgecolors="blue")
plt.plot(x,model.predict(x), color="red")
plt.show()
```

```python
## Define a deep linear neural network.
model = Sequential([
    Dense(4, input_shape=(1,), activation="linear"),
    Dense(1),
    Activation('linear')
])

## Prepare the model for training.
model.compile(loss='mean_squared_error',
              optimizer="nadam",
              metrics=['accuracy'])

## Train the model with data (x,y).
model.fit(x, y,
          epochs=20000,
          batch_size=128,
          verbose=False)

## Plot the figure and fit model.
f = plt.figure()
plt.scatter(x,y,facecolors="none",edgecolors="blue")
plt.plot(x,model.predict(x), color="red")
plt.show()
```

```python
## Define a deep non-linear (sigmoid) network.
model = Sequential([
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(1),
    Activation('linear')
])

## Prepare the model for training.
model.compile(loss='mean_squared_error', optimizer="nadam")

## Train the model.
model.fit(x, y,
          epochs=20000,
          batch_size=128,
          verbose=False)

## Plot the data and model fit.
f = plt.figure()
plt.scatter(x,y,facecolors="none",edgecolors="blue")
plt.plot(x,model.predict(x), color="red")
plt.show()
```

```python
model = Sequential([
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(1),
    Activation('linear')
])
model.compile(loss='mean_squared_error', optimizer="nadam")
model.fit(x, y,
          epochs=20000,
          batch_size=128,
          verbose=False)
f = plt.figure()
plt.scatter(x,y,facecolors="none",edgecolors="blue")
plt.plot(x,model.predict(x), color="red")
plt.show()
```

```python
model = Sequential([
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(1),
    Activation('linear')
])
model.compile(loss='mean_squared_error', optimizer="nadam")
model.fit(x, y,
          epochs=20000,
          batch_size=128,
          verbose=False)
f = plt.figure()
plt.scatter(x,y,facecolors="none",edgecolors="blue")
plt.plot(x,model.predict(x), color="red")
plt.show()
```

```python
## Define new data (fourth degree polynomial)
x = np.linspace(-1,3.5,75)
y = x*x*x*x - 5*x*x*x + 5*x*x + 5*x + 6

f = plt.figure()
plt.plot(x, y, 'o')

model = Sequential([
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(4, input_shape=(1,), activation="sigmoid"),
    Dense(1),
    Activation('linear')
])
model.compile(loss='mean_squared_error', optimizer="nadam")
model.fit(x, y, epochs=20000, batch_size=128, verbose=False)

f = plt.figure()
plt.scatter(x,y,facecolors="none",edgecolors="blue")
plt.plot(x,model.predict(x), color="red")
plt.show()
```
